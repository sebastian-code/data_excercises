```python
import time
import requests
import cloudscraper
from bs4 import BeautifulSoup
```

```python
url = "/novel/omniscient-readers-viewpoint-lightnovel-24041419?tab=chapters"
```

```python
def get_lightnovel_page(url):
    scraper = cloudscraper.create_scraper()
    page = scraper.get(f"https://www.lightnovelpub.com{url}")
    return BeautifulSoup(page.text, "html.parser")

def get_pagination_links(soup):
    links = [i["href"] for i in soup.find("ul", {"class": "pagination"}).find_all("a")]
    return [(f"{url}&page={count+2}&chorder=asc") for count, link in enumerate(links)]

def get_page_chapters(soup):
    chapters = soup.find("ul", {"class": "chapter-list"})
    return [i["href"] for i in chapters.find_all("a")]

def get_novel_chapters_links(links):
    novel_chapters_links: list = []
    for link in links:
        soup = get_lightnovel_page(link)
        page_chapters.extend(get_page_chapters(soup))
        time.sleep(60)
    
    return novel_chapters_links

def get_chapter_text(url):
    chapter = get_lightnovel_page(url)
    [p.decompose() for p in chapter("p", {"class": "dumckp"})]
    return [p.text for p in chapter("p")]
```

```python
soup = get_lightnovel_page(url)
```

```python
pagination_links = get_pagination_links(soup)
```

```python
novel_chapters: list = get_novel_chapters_links(pagination_links)
```

```python
for link in get_pagination_links(soup):
    soup = get_lightnovel_page(link)
    page_chapters.append(get_page_chapters(soup))
    time.sleep(60)

page_chapters
```
